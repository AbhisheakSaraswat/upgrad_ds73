```python
plt.figure(figsize = (6,4))
sns.countplot(data = df, y = 'PaymentMethod', hue = 'PaperlessBilling')
plt.title("Number of customer per payment method")
plt.ylabel("Payment method")
plt.xlabel("Count")
plt.legend(title = 'Paperless billing')
```

```python
# Plot the avg charges by payment method
```

```python
plt.figure(figsize = (6,4))
sns.barplot(data = df, y = 'PaymentMethod', x = 'TotalCharges',
           estimator='mean', errorbar=None)
plt.xlabel("Mean of Total Charges")
plt.ylabel("Payment method")
plt.title("Avg Charges per payment method.")
```

```python
# using a facet grid to plot bar graphs for avg charges by payment method
```

```python
bar_plot = sns.catplot(data = df, kind = 'bar', x = 'TotalCharges',
                      y = 'PaymentMethod', col = 'PaperlessBilling',
                      estimator='mean', errorbar=None)
bar_plot.set_axis_labels('Mean total charges','Payment method')
bar_plot.fig.suptitle('Avg total charges per payment method split by billing type', y = 1.03)
```

### Box Plot:

* 1. Minimum
* 2. Q1
* 3. Q2
* 4. Q3
* 5. Maxmium

```python
sns.boxplot(data = tips, x = 'day', y = 'total_bill')
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(data = df, x = 'tenure')
plt.title("Distribution of tenure")
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(data = df, x = 'PaymentMethod', y = 'tenure')
plt.title("Distribution of tenure accross different payment method")
plt.xlabel("Payment method")
plt.ylabel("Tenure")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(data = df, x = 'PaymentMethod', y = 'tenure',
           hue = 'PaperlessBilling')
plt.title("Distribution of tenure accross different payment method")
plt.xlabel("Payment method")
# plt.legend(loc = 'upper left', title = "Paperless Billing")
plt.legend(loc = 'upper left', bbox_to_anchor= (1,1))
plt.ylabel("Tenure")
plt.show()
```

```python
plt.figure(figsize = (7,3))
sns.swarmplot(x = 'day', y = 'total_bill', data = tips, palette="viridis")
plt.xlabel("Day of the week")
plt.ylabel("Total Bill ($)")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(data = tips, x = 'time', y = 'total_bill')
plt.title("Total bill distribution by meal time")
plt.xlabel("meal Time")
plt.ylabel("Total Bill")
plt.show()
```

```python
tips.head()
```

```python
plt.figure(figsize=(6,3))
sns.violinplot(x = 'day', y = 'total_bill', data = tips, palette="viridis")
plt.title("Total bill distribution by day of the week")
plt.xlabel("Day of the week")
plt.ylabel("Total bill ($)")
plt.show()
```

```python
plt.figure(figsize=(6,3))
sns.violinplot(x = 'day',
               y = 'total_bill',
               data = tips,
               hue = 'sex',
               palette="Set1",
              split=True)
plt.title("Total bill distribution by day of the week")
plt.xlabel("Day of the week")
plt.ylabel("Total bill ($)")
plt.legend(title = 'Gender')
plt.show()
```

```python
g = sns.relplot(x = 'total_bill',
               y = 'tip',
               data = tips,
               hue = 'time',
               col = 'day',
               row = 'sex',
               palette='Set1',
               height = 3, aspect=1.2)
g.set_titles(col_template = "{col_name} Day", row_template = "{row_name} Gender")
g.set_axis_labels("Total Bill ($)", "Tip ($)")
plt.suptitle("Rational scatter plot by day and gender")
plt.subplots_adjust(top=0.9)
plt.show()
```

### Seaborn

* Categorical Plots: Bar plot, count plot, cat plot, point plot etc
* Univariate Plots: box plot, dis plot, violin plot
* Bivariate plots (2 variables) :  scatter plot, regression plot
* Multivariate : pair plot, face grid, relational plot
* matrix plots: heatmap, cluster map
